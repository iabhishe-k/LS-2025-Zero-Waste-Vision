# Problem 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('images/image1.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
sharpened = cv2.filter2D(img, -1, kernel)


cv2.imshow("Original vs Sharpened", np.hstack((img, sharpened)))
cv2.waitKey(0)
cv2.destroyAllWindows()


sobelx = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=3)

cv2.imshow("Sobel X", sobelx)
cv2.imshow("Sobel Y", sobely)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Problem 2

```python
gray = cv2.imread('images/image1.jpg', 0)


plt.figure()
plt.title("Original Histogram")
plt.hist(gray.ravel(), 256, [0, 256])
plt.show()


equalized = cv2.equalizeHist(gray)


plt.figure()
plt.title("Equalized Histogram")
plt.hist(equalized.ravel(), 256, [0, 256])
plt.show()


cv2.imshow("Original", gray)
cv2.imshow("Equalized", equalized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Problem 3

```python
img = cv2.imread('images/image1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
edges = cv2.Canny(blurred, 100, 200)


combined = np.hstack((gray, blurred, edges))

cv2.imshow("Original | Blurred | Edges", combined)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Problem 4

```python
img = cv2.imread('images/image1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
bright = np.where(gray > 200)
count = 0
for y, x in zip(*bright):
    cv2.circle(img, (x, y), 1, (255, 255, 255), -1)
    count += 1

print("Number of pixels have a brightness >200= ", count)
cv2.imshow("Circle", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

img2 = cv2.imread('images/image1.jpg')
boxes = [(50, 60, 150, 120), (200, 180, 300, 250)]
for i, (x1, y1, x2, y2) in enumerate(boxes):
    cv2.rectangle(img2, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img2, f'Object {i+1}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
cv2.imshow("Annotate", img2)
cv2.waitKey(0)
cv2.destroyAllWindows()


canvas = np.zeros((500, 500, 3), dtype=np.uint8)
cv2.rectangle(canvas, (50, 50), (200, 150), (0, 0, 255), 2)
cv2.circle(canvas, (300, 300), 40, (0, 255, 0), -1)
cv2.line(canvas, (0, 0), (500, 500), (255, 0, 0), 2)
cv2.putText(canvas, "Abhishek", (370, 490), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

cv2.imwrite("problem4_output.jpg", canvas)
cv2.imshow("Canvas", canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Problem 5

```python

img = cv2.imread('images/image1.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)


t = np.hstack((h,s,v))
cv2.imshow("Hue | Saturation | Value", t)
cv2.waitKey(0)
cv2.destroyAllWindows()

reconverted = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

t = np.hstack((img, hsv, reconverted))
cv2.imshow("Original BGR | HSV | Reconverted BGR", t)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
